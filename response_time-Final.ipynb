```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('sample.csv',parse_dates=[1])
```

```python
df.sort_values(by=['Sensor_id','Timestamp'])
```

```python
sensors=list(df['Sensor_id'].unique())
sensors
full_at=100
empty_at=20
```

```python
for sensor in sensors:
    working=pd.DataFrame(columns=['Sensor_id', 'Timestamp', 'Quantity'])
    working=working.append(df[df['Sensor_id']==sensor],ignore_index=True)
    print(working)
    
```

```python

picked=pd.DataFrame()
for sensor in sensors:
    working=pd.DataFrame(columns=['Sensor_id', 'Timestamp', 'Quantity'])
    working=working.append(df[df['Sensor_id']==sensor],ignore_index=True)
    
    for i in working.index:
        if ((working.loc[i,'Quantity']<=empty_at) & 
            (working.loc[max(i-1,0),'Quantity']>empty_at)) | ((working.loc[i,'Quantity']==full_at) & 
                                                        (working.loc[max(i-1,0),'Quantity']<full_at)):
            picked=picked.append(working.loc[i,:],ignore_index=True)

picked=picked.sort_values(by=['Sensor_id','Timestamp'])
picked           
```

```python
response_time_df=pd.DataFrame()
for sensor in sensors:
    working2=pd.DataFrame(columns=[ 'Quantity','Sensor_id', 'Timestamp','Response_time'])
    working2=working2.append(picked[picked['Sensor_id']==sensor],ignore_index=True)
    for j in working2.index:
        if working2.loc[j,'Quantity']==full_at:
            working2.loc[j,'Response_time']=working2.loc[j,'Timestamp']-working2.loc[max(0,j-1),'Timestamp']
            response_time_df=response_time_df.append(working2.loc[j,:],ignore_index=True)
        else:
            working2.loc[j,'Response_time']=np.nan
            response_time_df=response_time_df.append(working2.loc[j,:],ignore_index=True)

response_time_df
```

```python
response_time_df.dropna(inplace=True)
```
